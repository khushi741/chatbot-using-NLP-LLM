```python
# importing necessary libraries
from transformers import AutoTokenizer
import gradio as gr
from transformers import pipeline
model = 'deepset/roberta-base-squad2'
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
nlp = pipeline("question-answering", model = model,tokenizer=tokenizer)
```

```python
context = '''My name is khushi and i love working on the LLMs aand Generative AIand i live in Delhi'''
```

```python
question = "what is my name?"

result = nlp(question = question,context = context)
result
```

```python
question = "what do i love?"

result = nlp(question = question,context = context)
result
```

```python
#creating the function
def func(context,question):
  result = nlp(question = question,context = context)
  return result['answer']
```

```python
#creating the interface
app = gr.Interface(fn=func,inputs=['textbox','text'],outputs = 'textbox',title = "khushi's Bot",theme = 'dark-grass')
```
